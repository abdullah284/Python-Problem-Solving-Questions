Madison is fond of toys. Her friend Mason works in a toy manufacturing factory . Mason has a 2D board A of size H x W with H rows and W columns. The board is divided into cells of size 1 x 1 with each cell indicated by its coordinate (i,j). The cell (i,j) has an integer A[ij] written on it. To create the toy Mason stacks A[ij] number of cubes of size 1 x 1 x 1 on the cell (i,j).

Given the description of the board showing the values of A[ij] and that the price of the toy is equal to the 3d surface area find the price of the toy.

Input Format:

*   A 2D array of size H x W representing the size of the board with each value representing the stacks.

Constraints:


*   H > 0
*   W < 101
*   0 < A[ij] < 101





Output Format:


*   Print the price of the toy.





Sample Input:


*   [[1]]
Sample Output:

*   6

Explanation:

The surface area of the cube 1 x 1 x 1 is 6.

Sample input:


*   [[1,3,4],

    [2,2,3],

    [1,2,4]]



Sample output:


*   60

Explanation:



cube2.png

The object is rotated so the front row matches column 1 of the input, heights 1, 2, and 1.



*   The front face is 1 + 2 + 1 = 4 units in area.

*   The top is 3 units.

*   The sides are 4 units.
*   None of the rear faces are exposed.


*   The underside is 3 units


The front row contributes 4 + 3 + 4 + 3 = 14 units to the surface area. If we do the same for the other rows, we will have a total of 60 units.



The front face is 1 + 2 + 1 = 4 units in area.
The top is 3 units.
The sides are 4 units.
None of the rear faces are exposed.
The underside is 3 units.
The front row contributes 4 + 3 + 4 + 3 = 14 units to the surface area.

```python

def surfaceArea(A):
  area_total = 0
  side_dif = 0
  top_bottom = (len(A) * len(A[0])) * 2

  for i in range(len(A)):
    for j in range(len(A[i])):
      if (j+1) < len(A[i]):
        if A[i][j] > A[i][j+1]:
          side_dif += A[i][j] - A[i][j+1]
      else:
        side_dif += A[i][j]
      if (j-1) >= 0:
        if  A[i][j] > A[i][j-1]:
          side_dif += (A[i][j] - A[i][j-1])
      else:
        side_dif += A[i][j]
      if (i+1) < len(A):
        if A[i][j] > A[i+1][j]:
          side_dif += (A[i][j] - A[i+1][j])
      else:
        side_dif += A[i][j]
      if (i-1) >= 0:
        if A[i][j] > A[i-1][j]:
          side_dif += (A[i][j] - A[i-1][j])
      else:
        side_dif += A[i][j]
      area_total = area_total+ side_dif
      side_dif = 0


  area_total += top_bottom



  return (area_total)
```

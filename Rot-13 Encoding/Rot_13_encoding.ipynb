Write a program to read the contents of a file using the Rot-13 encode

rot-13.png

Assumption: The input only contains upper case letters and no numbers or special characters










```python
def rot13(input):

  letters = list(input)

  new_string = ""

  for i in range(len(letters)):
    if letters[i] == " ":
      new_string = new_string + " "
      continue
    index = ord(letters[i])
    if index <= 77:
      new_index = index + 13
      new_letter = chr(new_index)
      new_string = new_string + new_letter
    elif index >= 78:
      new_index = index - 13
      new_letter = chr(new_index)
      new_string = new_string + new_letter


  print(new_string)

```

```python
rot13("URYYB JBEYQ")
```
